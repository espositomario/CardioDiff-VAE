```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import gseapy as gp
from gseapy import barplot, dotplot
import pickle        

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
plt.rcParams['savefig.dpi'] = 250
plt.rcParams['figure.dpi'] = 36
plt.rcParams['axes.facecolor'] = 'white'
%matplotlib inline
```

### ORA (Over representation analysis) for each subset of VARIABLE genes divided by the CTmax

```python
DIR_DATA = f'../data/RNA_CV/BOTTOM4000/'
DIR_FIG = DIR_DATA
with open(f'{DIR_DATA}dict.pkl', 'rb') as f:
    BOTTOM_CV = pickle.load(f)
```

```python
STABLE = [gene for gene_list in BOTTOM_CV.values() for gene in gene_list]
```

```python

DIR_DATA = f'../data/RNA_CV/TOP4000/'
DIR_FIG = DIR_DATA
with open(f'{DIR_DATA}dict.pkl', 'rb') as f:
    GENE_CLUSTERS = pickle.load(f)
GENE_CLUSTERS['STABLE'] = STABLE
```

```python

#
RESULTS = {}
for CT, gene_list in GENE_CLUSTERS.items():
    enr = gp.enrichr(gene_list= gene_list,
                    #background=  f'{DIR_DATA}/background.list',
                    gene_sets=['KEGG_2019_Mouse',
                                'WikiPathways_2019_Mouse',
                                'GO_Biological_Process_2023',
                                #'GO_Molecular_Function_2023'
                                ],
                    organism='Mouse', 
                    outdir=None, 
                    verbose=True )

    RESULTS[CT]= {'result': enr.results, 'n_genes':len(gene_list)}

#### Remove from Term final string after '('
RESULTS_CLEANED = RESULTS.copy()
for key,value in RESULTS_CLEANED.items():
        x=RESULTS_CLEANED[key]['result']['Term'].str
        NEW_COL=[]
        for el in list(x.split('(')):
                NEW_COL.append(el[0])
        RESULTS_CLEANED[key]['result']['Term']=NEW_COL

```

```python
COL_DICT = {
    'KEGG_2019_Mouse': '#66C2A5',
    'WikiPathways_2019_Mouse': '#FC8D62',
    'GO_Biological_Process_2023': '#8DA0CB',
    'GO_Molecular_Function_2023': '#E78AC3'
}
```

```python
# 
for CT, value in RESULTS_CLEANED.items():
    n_genes=RESULTS_CLEANED[CT]['n_genes']
    # count top sig to scale dots
    df = RESULTS_CLEANED[CT]['result']
    gene_set_counts = df[df['Adjusted P-value'] < 0.05]['Gene_set'].value_counts().values
    gene_set_counts[gene_set_counts>5] = 5
    top_sig = gene_set_counts.sum()
    print(top_sig)
    try:
        barplot(
            RESULTS_CLEANED[CT]['result'],
            column="Adjusted P-value",
            group='Gene_set',
            color=COL_DICT,
            top_term=5,
            title=f'{CT}, n={n_genes}',
        )
        plt.savefig(f'{DIR_FIG}/{CT}.pdf', format="pdf", bbox_inches="tight")

    except ValueError:
        with open(f'{DIR_FIG}/{CT}_no_sign_terms.error', "w") as file:
            file.write("No significant terms")

# Adjust layout
plt.tight_layout()
# Show the plot
plt.show()
```
