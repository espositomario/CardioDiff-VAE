### Map Gene names Human2Mouse (Manually AAA123A-> Aaa123a)

```python
def hum2mouse(gene):
	new_gene = gene[0] + gene[1:].lower()
	return new_gene
```

```python
GENE_SETS_low = {}

for term, genes in GENE_SETS.items():
    new_genes = []
    for gene in genes: 
            new_gene = hum2mouse(gene)
            new_genes.append(new_gene)
    GENE_SETS_low[term] = new_genes
```
